## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##  Data Loading

The dataset is loaded using pandas. Since the dataset contains encoding and delimiter issues, appropriate parameters are used to correctly parse the file.


```python
df = pd.read_csv(
    r"C:\Users\akhia\Documents\IBM_data_cleaning\data\messy_IMDB_dataset.csv",
    sep=';',
    encoding='latin1'
)
df.head()
```

## Data Cleaning

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.head()
```

##  Column Name Cleaning

Column names are cleaned to ensure consistency by removing extra spaces, converting to lowercase, and standardizing naming conventions.


```python
df.rename(columns={'imbd_title_id' : 'imdb_title_id','original_titlÃª' : 'original_title', 'genrÃ«Â¨' : 'genre'},inplace=True
)
```

##  Data Issues Identified

The dataset contains several inconsistencies:

* Incorrect delimiter causing parsing issues
* Encoding problems leading to corrupted text
* Numerical columns containing symbols and invalid characters
* Missing values in multiple columns
* Inconsistent categorical values
* Presence of malformed or empty rows

These issues need to be addressed before performing any analysis.


```python
df.drop(columns=['unnamed:_8'],inplace=True)
```

```python
df.dropna(how='all', inplace=True)
```

```python
import numpy as np

df.replace(
    ['Nan', 'NULL', '-', 'Inf', 'Not Applicable', '#N/A', ''],
    np.nan,
    inplace=True
)
```

```python
df.head()
```

```python
df['income'] = (df['income'].str.replace('$','' , regex=False).str.replace(',' , '' , regex=False).str.replace('o', '',regex=False).str.strip())

df['income'] = pd.to_numeric(df['income'], errors='coerce')
```

```python
df['votes'] = df['votes'].str.replace('.', '' , regex=False).str.strip()
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df['score'] = (
    df['score']
    .str.replace(',', '.', regex=False)
    .str.replace('f', '', regex=False)
    .str.replace('..', '.', regex=False)
    .str.strip()
)

df['score'] = pd.to_numeric(df['score'], errors='coerce')
```

```python
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
```

```python
df['country'] = df['country'].str.replace('.','',regex=False).str.strip()

df['country'] = df['country'].replace({
    'US': 'USA',
    'US.': 'USA'
})
```

```python
df['genre'] = df['genre'].str.replace('\t', ',', regex=True)
df['genre'] = df['genre'].str.strip()
```

```python
df['release_year'] = pd.to_datetime(df['release_year'], errors='coerce')
```

##  Dataset Overview

We analyze the dataset structure, data types, and summary statistics to identify potential issues such as incorrect data types and missing values.


```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['release_year'])
```

##  Handling Missing Values

Missing values are handled using appropriate strategies:

* Numerical columns â filled using median
* Categorical columns â filled with placeholder values
* Critical fields â rows removed where necessary


```python
df['duration'].fillna(df['duration'].median(), inplace=True)
```

```python
df['score'].fillna(df['score'].median(), inplace=True)
```

##  Categorical Data Standardization

Categorical columns contain inconsistent labels due to typos and formatting differences.
These values are standardized to ensure uniform representation across the dataset.


```python
df['content_rating'].fillna('Unknown', inplace=True)
```
